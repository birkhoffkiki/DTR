```python
import os
import torch
from PIL import Image
from torchvision.transforms import functional as F
import Attention_GAN
import matplotlib.pyplot as plt

os.environ['CUDA_VISIBLE_DEVICES']='0'
device = torch.device('cuda')
```

```python
# AF2HE Demo
# build generator and discriminator
netG = Attention_GAN.Generator(n_channels=64, in_channels=3, batch_norm=False, 
    out_channels=3, padding=1, pooling_mode="maxpool",).to(device)
msg = netG.load_state_dict(torch.load('weights/af2he_weight.pth', map_location='cpu'), strict=True)
print(msg)

input_image = Image.open('assets/af2he/input.png').convert('RGB')
target_image = Image.open('assets/af2he/target.png').convert('RGB')
# to tensor -1 to 1
input_img = (F.to_tensor(input_image).unsqueeze(0).to(device) - 0.5) *2
target_img = (F.to_tensor(target_image).unsqueeze(0).to(device) - 0.5) *2

# inference
with torch.no_grad():
    netG.eval()
    output_img = netG(input_img)
    output_img = (output_img.squeeze(0).permute(1, 2, 0).cpu().numpy() + 1)/2
    output_img = (output_img * 255).astype('uint8')
# show image
plt.subplot(1, 3, 1)
plt.imshow(input_img.squeeze(0).permute(1, 2, 0).cpu().numpy() * 0.5 + 0.5)
plt.title('Input Image')
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(target_img.squeeze(0).permute(1, 2, 0).cpu().numpy() * 0.5 + 0.5)
plt.title('Target Image')
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(output_img)
plt.axis('off')
plt.title('Output Image')
```
